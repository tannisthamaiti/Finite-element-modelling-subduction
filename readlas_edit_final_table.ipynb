```python
import numpy as np
import pandas as pd
```

```python
listOfInt = []
with open('15_9-19_BT2_CPI.las') as f:
    for i, curline in enumerate(f):
        if curline.startswith("#"):
            numstart = i
with open('15_9-19_BT2_CPI.las') as f:
    lines_after_17 = f.readlines()[numstart+1:]
listOfStr = lines_after_17[0].split()[1:]
numlines = len(lines_after_17)
for i in range(1,numlines):
     listOfInt.append(lines_after_17[i].split())
dfObj = pd.DataFrame(listOfInt) 
dfObj.columns = listOfStr
dfObj.head()
```

```python
dfxls = pd.read_excel('15_9-19 BT2 Facies.xlsx')
dfxls[dfxls.columns[6]] = dfxls[dfxls.columns[6]].replace({'UNKNOWN': np.nan})
dfxls[dfxls.columns[6]].fillna(method = 'bfill', inplace = True)
dfxls.head()
```

```python
class_lith =[]
top_depth = dfxls[dfxls.columns[4]]
for i in range(0,len(dfObj)): #
    top_depth_diff = abs(float(dfObj[dfObj.columns[0]][i])-top_depth)
    minval = min(top_depth_diff)
    indexes = [i for i,x in enumerate(top_depth_diff) if x == minval]
    class_lith.append(dfxls[dfxls.columns[6]][indexes[0]])
```

```python
dfObj['Class_lith'] = np.nan
dfObj['Class_lith']= class_lith
```
